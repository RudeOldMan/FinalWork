**(4) Consider a house with an electric heater. How the electricity consumption of the heater would be affected by the air temperature? Is there a way to change this?**

Answer:Electric heater is set to certain heating level with thermostat and it measures the air temperature with a sensor. If air around the heater is colder than desired level, device heats more and consumes more electricity. This could happen even if heat level in the room is sufficient.

Ways to change this:
-pick suitable place for heater, not in a drafty place
-avoid unnecessary openings of door or window
-if you have some rooms that don’t require extra heating at the moment, close the doors
